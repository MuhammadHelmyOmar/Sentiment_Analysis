### Text Preprocessing

```python
# !pip install farasapy
```

```python
from nltk.tokenize import word_tokenize
import re # Regex
import string # To remove punctuations
from farasa.stemmer import FarasaStemmer
```

```python
def removeStopwords(tokens):
    tokens = tokens.split()
    return ' '.join([w for w in tokens if (w not in arab_stopwords) and (len(w) >= 2)])
```

```python
def normalize(text):
    # Remove three or more repetitions of a character
    pattern = re.compile(r"(.)\1{2,}",re.DOTALL)
    text = pattern.sub(r"\1\1",text)
    # Normalize alif
    text = text.replace(u"\u0625", u"\u0627")  # HAMZA below, with LETTER ALEF
    text = text.replace(u"\u0622", u"\u0627")  # ALEF WITH MADDA ABOVE, with LETTER ALEF
    text = text.replace(u"\u0623", u"\u0627")  # ALEF WITH HAMZA ABOVE, with LETTER ALEF
    # Normalize taa
    text = text.replace(u"\u0629", u"\u0647") # taa' marbuuTa, with haa'
    # Normalize yaa
    text = text.replace(u"\u064A", u"\u0649")  # yaa' with 'alif maqSuura
    # Remove diacritics 
    text = text.replace(u"\u064B", "")  # fatHatayn
    text = text.replace(u"\u064C", "")  # Dammatayn
    text = text.replace(u"\u064D", "")  # kasratayn
    text = text.replace(u"\u064E", "")  # fatHa
    text = text.replace(u"\u064F", "")  # Damma
    text = text.replace(u"\u0650", "")  # kasra
    text = text.replace(u"\u0651", "")  # shaddah
    text = text.replace(u"\u0652", "")  # sukuun
    return text
```

```python
punctuations = [word.strip() for word in string.punctuation] + ['؛', 'ـ','،', '؟', '”', '…', '“', '–' ]
print(len(punctuations))

def removePunct(text):
    clean_text = ''
    for c in text:
        if c not in punctuations:
            clean_text += c
    return clean_text
    
```

```python
# Noise Removal
def removeNoise(text):
    # Remove extra whitespace
    text = re.sub('\s+', ' ', text)
    # Remove numbers
    text = re.sub('\d+', ' ', text)
    return text
```

```python
prep_data = pd.DataFrame()
```

```python
# Applying cleaning functions
prep_data['clean_question'] = data['Question'].apply(lambda x : removeStopwords(x)).apply( lambda x : normalize(x)).apply( lambda x : removePunct(x)).apply( lambda x : removeNoise(x))
prep_data.head()
```

### Training Word2Vec

```python
from gensim.models import Word2Vec
import multiprocessing
from time import time
from gensim.models.phrases import Phraser, Phrases
```

```python
sent = [row for row in prep_data.tokenized_question]
phrases = Phrases(sent, min_count= 1, progress_per= 50000)
bigram = Phraser(phrases)
sentences = bigram[sent]
print(phrases.vocab)
sentences[1]
```

```python
bigram.phrasegrams
```

```python
w2v = Word2Vec(
    min_count = 2,
    window = 4,
    vector_size = 300,
    negative = 20,
    alpha = 0.03,
    min_alpha= 0.0007,
    workers= multiprocessing.cpu_count() - 1
)

start_time = time()
w2v.build_vocab(sentences, progress_per= 50000)
print('Time to build vocab: {} mins'.format( round( ( time() - start_time ) / 60, 2 ) ) )
```

```python
start_time = time()
w2v.train(sentences, total_examples= w2v.corpus_count, epochs = 30, report_delay=1)
print('Time to train the model: {} mins'.format( round( ( time() - start_time ) / 60, 2 ) ) )
```

```python
w2v.init_sims(replace=True)
w2v.save("word2vec.model")
```

```python
# Replacing words in sentences with their tfidf scores
def create_tfidf_dictionary(x, transformed_data, features):
    '''
    create dictionary for each input sentence x, where each word has assigned its tfidf score
    x - row of dataframe, containing sentences, and their indexes,
    transformed_data - all sentences transformed with TfidfVectorizer
    features - names of all words in corpus used in TfidfVectorizer
    '''
    vector_coo = transformed_data[x.name].tocoo()
    vector_coo.col = features.iloc[vector_coo.col].values
    dict_from_coo = dict(zip(vector_coo.col, vector_coo.data))
    return dict_from_coo
```

```python
def replace_tfidf_words(x, transformed_data, features):
    '''
    replacing each word with it's calculated tfidf dictionary with scores of each word
    x - row of dataframe, containing sentences, and their indexes,
    transformed_data - all sentences transformed with TfidfVectorizer
    features - names of all words in corpus used in TfidfVectorizer
    '''
    dictionary = create_tfidf_dictionary(x, transformed_data, features)   
    return list(map(lambda y:dictionary[f'{y}'], x.clean_question.split()))
```
